## Step 6: Data Visualisation

```python
print(data.columns)

# To create a Dataframe object to hold the data in a tabular format
df = pd.DataFrame(data.head(20))
```

```python
# Generate a pie chart to visualize the distribution of values in the 'Country' column
df['Country'].value_counts().plot(kind='pie', autopct='%1.1f%%')

# Add a title to the pie chart indicating the variable being visualized
plt.title("Country")

# Display the pie chart
plt.show()
```

```python
# Generate a bar chart to visualize the distribution of values in the 'NumDealsPurchases' column
df['NumDealsPurchases'].value_counts().plot(kind='bar', color='green')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Number of Deal Purchases')

# Add a label to the y-axis indicating the count of each value
plt.ylabel('Count')

# Add a title to the bar chart indicating the purpose of the visualization
plt.title('Distribution of Number of Deal Purchases')

# Display the bar chart
plt.show()
```

```python
# Generate a horizontal bar chart to visualize the distribution of values in the 'Education' column
df['Education'].value_counts().plot(kind='barh', color='yellow')

# Add a label to the x-axis indicating the count of each value
plt.xlabel('Count')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('Education Level')

# Add a title to the horizontal bar chart indicating the purpose of the visualization
plt.title('Distribution of Education Levels')

# Display the horizontal bar chart
plt.show()
```

```python
# Convert the 'Dt_Customer' column to datetime format if it's not already in datetime format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Sort the DataFrame by the 'Dt_Customer' column to ensure the data is plotted in chronological order
df.sort_values(by='Dt_Customer', inplace=True)

# Plot a line chart to visualize the trend of 'Income' over time
plt.plot(df['Dt_Customer'], df['Income'], color='red', marker='o', linestyle='-')

# Add a label to the x-axis indicating the date of customer registration
plt.xlabel('Date of Customer Registration')

# Add a label to the y-axis indicating the income
plt.ylabel('Income')

# Add a title to the line chart indicating the purpose of the visualization
plt.title('Trend of Income Over Time')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Add grid lines to the plot for better visualization
plt.grid(True)

# Display the line chart
plt.show()
```

```python
# Generate a histogram to visualize the distribution of values in the 'Income' column
plt.hist(df['Income'], bins=20, color='green', edgecolor='black')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Income')

# Add a label to the y-axis indicating the frequency of observations
plt.ylabel('Frequency')

# Add a title to the histogram indicating the purpose of the visualization
plt.title('Distribution of Income')

# Add grid lines to the histogram for better visualization
plt.grid(True)

# Display the histogram
plt.show()
```

```python
# Generate a scatter plot to visualize the relationship between 'Income' and 'NumWebPurchases'
plt.scatter(df['Income'], df['NumWebPurchases'], color='blue', alpha=0.5)

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Income')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('Number of Web Purchases')

# Add a title to the scatter plot indicating the purpose of the visualization
plt.title('Scatter Plot: Income vs. Number of Web Purchases')

# Add grid lines to the scatter plot for better visualization
plt.grid(True)

# Display the scatter plot
plt.show()
```
