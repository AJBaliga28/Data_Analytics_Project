## Step 6: Data Visualisation

```python
print(data.columns)

# To create a Dataframe object to hold the data in a tabular format
df = pd.DataFrame(data)
```

```python
# Sum up the spending across all product categories for each customer
data['Total_Spending'] = data[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Segment customers based on their total spending
spending_bins = [0, 1000, 2000, 3000, 4000, float('inf')]
spending_labels = ['0-1000', '1001-2000', '2001-3000', '3001-4000', '4001+']
data['Spending_Category'] = pd.cut(data['Total_Spending'], bins=spending_bins, labels=spending_labels)

# Count the number of customers in each spending category
spending_distribution = data['Spending_Category'].value_counts()

# Plot the distribution of spending categories using a pie chart
plt.figure(figsize=(8, 6))
plt.pie(spending_distribution, labels=spending_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Customer Spending Categories')

# Adjust the position of the legend to avoid overlapping labels
plt.legend(loc="best")

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Calculate the average number of purchases made through different channels
engagement_by_channel = data[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].mean()

# Visualize the average number of purchases across different channels using a bar chart
plt.figure(figsize=(8, 6))
engagement_by_channel.plot(kind='bar', color=['lightgreen', 'lightblue', 'lightcoral'])
plt.title('Average Number of Purchases Across Channels')
plt.xlabel('Channel')
plt.ylabel('Average Number of Purchases')
plt.xticks(rotation=0)
plt.show()
```

```python
import datetime

# Calculate the current year
current_year = datetime.datetime.now().year

# Calculate the age of each customer
data['Age'] = current_year - data['Year_Birth']

# Define age groups
age_bins = [0, 30, 40, 50, 60, float('inf')]
age_labels = ['<30', '30-39', '40-49', '50-59', '60+']

# Categorize customers into age groups
data['Age_Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)

# Calculate the average engagement of customers for each age group
engagement_by_age_group = data.groupby('Age_Group', observed=False)[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].mean()

# Visualize the average engagement of customers across different age groups using a horizontal bar chart
plt.figure(figsize=(10, 6))
engagement_by_age_group.plot(kind='barh', stacked=True, colormap='viridis')
plt.title('Average Customer Engagement Across Age Groups')
plt.xlabel('Average Number of Purchases')
plt.ylabel('Age Group')
plt.legend(title='Channel')
plt.tight_layout()  # Adjust layout to move labels to the side
plt.show()
```

```python
# Convert 'Dt_Customer' column to datetime
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Group the data by month and calculate the average number of web visits
avg_web_visits_monthly = df.groupby(df['Dt_Customer'].dt.to_period('M'))['NumWebVisitsMonth'].sum()

# Plot the average number of web visits over time
plt.figure(figsize=(10, 6))
avg_web_visits_monthly.plot(marker='o', linestyle='-')
plt.title('Total Number of Web Visits per Month Over Time')
plt.xlabel('Month')
plt.ylabel('Total Number of Web Visits')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Extracting columns related to campaign responses
campaign_response_columns = ['ID', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
campaign_response_data = data[campaign_response_columns]

# Create an empty DataFrame to store the unique customer IDs who accepted the offer for each campaign
unique_customers_per_campaign = pd.DataFrame()

# Loop through each campaign column
for campaign in campaign_response_columns[1:7]:
    # Filter the data for customers who accepted the offer in the current campaign
    accepted_customers_campaign = campaign_response_data[campaign_response_data[campaign] == 1]
    
    # Create an empty list to store unique customer IDs for the current campaign
    unique_customers_campaign = []
    
    # Iterate over each row in the filtered data
    for index, row in accepted_customers_campaign.iterrows():
        # Check if the customer ID is already in the list of unique customer IDs
        if row['ID'] not in unique_customers_campaign:
            # Add the customer ID to the list
            unique_customers_campaign.append(row['ID'])
    
    # Count the number of unique customers who accepted the offer in the current campaign
    num_unique_customers_campaign = len(unique_customers_campaign)
    
    # Store the number of unique customers for the current campaign in the DataFrame
    unique_customers_per_campaign[campaign] = [num_unique_customers_campaign]

print(unique_customers_per_campaign)
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(unique_customers_per_campaign.columns, unique_customers_per_campaign.values[0], color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Number of Unique Customers Accepted per Campaign')
plt.xlabel('Campaign')
plt.ylabel('Number of Unique Customers Accepted')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Define the variables for the scatter plot
x = data['MntFruits']
y = data['MntSweetProducts']

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Scatter Plot of Amount Spent on Fruits vs Sweets')
plt.xlabel('Amount Spent on Fruits')
plt.ylabel('Amount Spent on Sweets')
plt.grid(True)
plt.show()
```
